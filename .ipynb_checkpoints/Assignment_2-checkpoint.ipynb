Normalize the data usitg the StandardScaler()

```python
scaled_features= StandardScaler()
```

```python
for col in hepatis.columns:
    if not (col== 'label'):
        hepatis[col]= scaled_features.fit_transform(hepatis[col].values.reshape(-1, 1))
```

```python
hepatis.head(3)
```

```python
X_train, X_test, y_train, y_test=train_test_split(hepatis.iloc[:, :-1], hepatis.iloc[:, -1], test_size=0.2, random_state=42)
```

Split the dataset into training and test sets

```python
Num_of_trials=10
```

```python
classifiers= [
    ('LR', LogisticRegression()),
    ('GNB', GaussianNB()),
    ('kNN', KNeighborsClassifier()),
    ('LDA', LinearDiscriminantAnalysis()),
    ('SVM', SVC(probability=True))]

param_grid= [
    {'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]},
    {'classifier__var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]},
    {'classifier__n_neighbors': [1, 3, 5, 7, 9],
     'classifier__weights': ['uniform', 'distance'],
     'classifier__algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']},
    {'classifier__solver': ['svd', 'lsqr', 'eigen']},
    {'classifier__kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
     'classifier__decision_function_shape': [ 'ovo', 'ovr'],
     'classifier__C': [0.01, 0.1, 1, 10, 100] }]
```

```python
scorer_auc= make_scorer(roc_auc_score)
scorer_mcc= make_scorer(matthews_corrcoef)
scorer_bas= make_scorer(balanced_accuracy_score)
scorer_f1= make_scorer(f1_score)
scorer_rec= make_scorer(recall_score)
scorer_pres= make_scorer(precision_score)
scorer_acu= make_scorer(accuracy_score)
scorer_avp= make_scorer(average_precision_score)


metrics= ['MCC', 'Balanced_accuracy', 'F1', 'Recall', 'Precision', 
         'Accuracy', 'AUC', 'Average_precision', 'Iteration', 'Model']


results= pd.DataFrame(columns=metrics)
results_test= pd.DataFrame(columns=metrics)
```

```python
start= time.time()
```

```python
for j in range(len(classifiers)):
    clf = classifiers[j]
    paramet = param_grid[j]
    pipe = Pipeline(steps=[('classifier', clf[1])])
    for i in range(Num_of_trials):
        outer_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=i)
        inner_cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=i)
        model = GridSearchCV(pipe, param_grid=paramet, cv=inner_cv, scoring=scorer_auc, refit=True)
        model.fit(X_train, y_train)
        best_model = model.best_estimator_
        model= clf[0]
        MCC= cross_val_score(best_model, X= X_train, y= y_train, cv= outer_cv, scoring= scorer_mcc)
        Balanced_accuracy= cross_val_score(best_model, X= X_train, y=  y_train, cv= outer_cv, scoring=scorer_bas)
        F1= cross_val_score(best_model, X= X_train, y= y_train, cv= outer_cv, scoring= scorer_f1)
        Recall= cross_val_score(best_model, X= X_train, y= y_train, cv= outer_cv, scoring= scorer_rec)
        Precision= cross_val_score(best_model, X= X_train, y= y_train, cv= outer_cv, scoring= scorer_pres)
        Accuracy= cross_val_score(best_model, X= X_train, y= y_train, cv= outer_cv, scoring= scorer_acu)
        AUC= cross_val_score(best_model, X= X_train, y= y_train, cv=outer_cv, scoring= scorer_auc)
        average_precision= cross_val_score(best_model, X= X_train, y= y_train, cv= outer_cv, scoring= scorer_avp)
        Iteration= i+1
        results.loc[len(results)]= [np.mean(MCC), np.mean(Balanced_accuracy), np.mean(F1),
                                      np.mean(Recall), np.mean(Precision), np.mean(Accuracy), np.mean(AUC),
                                      np.mean(average_precision), Iteration,
                                      model]
        y_pred = best_model.predict(X_test)
        test_MCC = matthews_corrcoef(y_test, y_pred)
        test_Balanced_accuracy = balanced_accuracy_score(y_test, y_pred)
        test_F1 = f1_score(y_test, y_pred)
        test_Recall = recall_score(y_test, y_pred)
        test_Precision = precision_score(y_test, y_pred)
        test_Accuracy = accuracy_score(y_test, y_pred)
        test_AUC = roc_auc_score(y_test, y_pred)
        test_average_precision = average_precision_score(y_test, y_pred)
        results_test.loc[len(results_test)] = [test_MCC, test_Balanced_accuracy, test_F1, test_Recall,
                                                test_Precision, test_Accuracy, test_AUC, test_average_precision,
                                                Iteration, model]
```

```python
end = time.time()
```

```python
time=end-start
time
```
